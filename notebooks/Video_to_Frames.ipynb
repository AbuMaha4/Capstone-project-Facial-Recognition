```python
from facenet_pytorch import MTCNN, extract_face
from PIL import Image, ImageDraw
import cv2
import pandas as pd
import numpy as np
```

```python
cap = cv2.VideoCapture("/Users/ben/neuefische/capstone/ai-with-python-series/07. Emotion Recognition using Live Video/content/Baseball Training.mp4")
success = True
count = 0
list_of_images_pil = []
while success:
    success, frame = cap.read()
    if success == False:
        break
    cv2_image = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    pil_image = Image.fromarray(cv2_image)
    list_of_images_pil.append(pil_image)
print(f'{len(list_of_images_pil)} images put into list_of_images_pil')
```
